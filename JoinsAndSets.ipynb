## JOINS

#### LEFT JOINS

#### 
Combine the Geographic_Location table with the Economic_Indicators table based on the Country_name column.

```python
%%sql

SELECT
    *
FROM
   geographic_location AS geo 
LEFT JOIN economic_indicators AS econ 
        ON geo.country_name=econ.country_name
LIMIT 5; 
```

###
Combine the previously joined tables with the Basic_Services table, again based on the Country_name column. We add an additional condition based on the Time_period column to make sure time periods align.

```python
%%sql

SELECT
    *
FROM
   geographic_location AS geo 
LEFT JOIN economic_indicators AS econ 
        ON geo.country_name=econ.country_name
LEFT JOIN basic_services AS servs
        ON geo.country_name=servs.country_name
        AND econ.Time_period=servs.time_period
    LIMIT 5; 
```

#### EX 1
Sometimes artists add a title track to their albums. This is a track that has the same title as the album. Write a query that returns albums that have a title track. Return rows in the `AlbumId` column from the `albums` table, `Title` from the `albums` table, and the `Name` from the `tracks` table, where the `album` table `Title` column matches the `tracks` table `Name` column.


```python
%%sql

SELECT
    albums.albumid,
    albums.title,
    tracks.name
FROM
    albums
JOIN tracks
        ON albums.albumid=tracks.albumid
WHERE
    tracks.name=albums.title
LIMIT 10;
```

#### EX 2
Suppose that in the previous exercise, we were additionally interested in knowing who the artists of the listed albums are. Write a query that can achieve this.

```python
%%sql

SELECT
    albums.albumid,
    albums.title,
    tracks.name,
    artists.name
FROM
    albums
INNER JOIN tracks
        ON albums.albumid=tracks.albumid
INNER JOIN artists
        ON albums.artistid=artists.artistid
WHERE
    tracks.name=albums.title
LIMIT 10;
```

#### EX 3
One use case for a LEFT JOIN is that it can be used to check for missing information. In this case, try to find out what media items have not been bought yet (i.e. are not an item in any invoice).

```python
%%sql

SELECT
    tracks.trackid,
    tracks.name,
    invoice_items.invoiceid
FROM 
    tracks
LEFT JOIN invoice_items
        ON tracks.trackid = invoice_items.trackid
LIMIT 10;
```

#### EX 4
In the results, the tracks that have a value of None (i.e. NULL) for InvoiceId are the ones that have not been purchased yet. Add a WHERE clause to only focus on these 'unpopular' tracks.

```python
%%sql

SELECT
    tracks.trackid,
    tracks.name,
    invoice_items.invoiceid
FROM 
    tracks
LEFT JOIN invoice_items
        ON tracks.trackid = invoice_items.trackid
WHERE invoice_items.invoiceid IS NULL
LIMIT 10;
```

#### Exercise 5
Let's suppose that, as part of a new business strategy, Chinook wants to develop new product categories for their media items that are based on genre and media type. Write a query that will list all possible product categories (i.e. all possible genre and media type combinations).

```python
%%sql

SELECT
    genres.Name AS 'Genre',
    media_types.Name AS 'Media Types'
FROM
    genres
CROSS JOIN
    media_types
LIMIT 10;
```

## UNION

### 
Suppose we want a summary of estimated unemployment rates per country for each given time period, and we expect that certain data entries of the unemployment rates might be missing.

To address this issue, we intend to replace the absent values by using the regional estimated unemployment rates given in order to enhance the completeness of our summary.

#### 1. Join tables
Obtain Time_period and Pct_unemployment columns for each country by combining the Geographic_Location and Economic_Indicators tables based on the Country_name for the first region, central and southern Asia.

```python
%%sql

SELECT
    geo.country_name,
    geo.region,
    econ.time_period,
    econ.pct_unemployment
FROM
    geographic_location AS geo
LEFT JOIN economic_indicators AS econ
        ON geo.country_name = econ.country_name 
WHERE region LIKE '%Central and Southern Asia%'
limit 10;
```

#### 2. Impute NULL values
To enhance the completeness of our summary, we'll address missing unemployment rate values. If any unemployment rate is absent, we'll replace it with the regional data, which is 19.59 in this case. We will then save this column as Pct_unemployment_imputed.

```python
%%sql

SELECT
    geo.country_name,
    geo.region,
    econ.time_period,
    IFNULL(econ.pct_unemployment,19.59) AS pct_unemployment_imputed
FROM
    geographic_location AS geo
LEFT JOIN economic_indicators AS econ
        ON geo.country_name = econ.country_name 
WHERE region LIKE '%Central and Southern Asia%'

UNION

SELECT
    geo.country_name,
    geo.region,
    econ.time_period,
    IFNULL(econ.pct_unemployment,22.64) AS pct_unemployment_imputed
FROM
    geographic_location AS geo
LEFT JOIN economic_indicators AS econ
        ON geo.country_name = econ.country_name 
WHERE region LIKE '%Eastern and South-Eastern Asia%'

UNION

SELECT
    geo.country_name,
    geo.region,
    econ.time_period,
    IFNULL(econ.pct_unemployment,24.43) AS pct_unemployment_imputed
FROM
    geographic_location AS geo
LEFT JOIN economic_indicators AS econ
        ON geo.country_name = econ.country_name 
WHERE region LIKE '%Europe and Northern America%'

UNION

SELECT
    geo.country_name,
    geo.region,
    econ.time_period,
    IFNULL(econ.pct_unemployment,24.23) AS pct_unemployment_imputed
FROM
    geographic_location AS geo
LEFT JOIN economic_indicators AS econ
        ON geo.country_name = econ.country_name 
WHERE region LIKE '%Latin America and the Caribbean%'

UNION

SELECT
    geo.country_name,
    geo.region,
    econ.time_period,
    IFNULL(econ.pct_unemployment,4.98) AS pct_unemployment_imputed
FROM
    geographic_location AS geo
LEFT JOIN economic_indicators AS econ
        ON geo.country_name = econ.country_name 
WHERE region LIKE '%Oceania%'

UNION

SELECT
    geo.country_name,
    geo.region,
    econ.time_period,
    IFNULL(econ.pct_unemployment,33.65) AS pct_unemployment_imputed
FROM
    geographic_location AS geo
LEFT JOIN economic_indicators AS econ
        ON geo.country_name = econ.country_name 
WHERE region LIKE '%Sub-Saharan Africa%';
```

#### EX 2
In some cases, artists will create eponymous albums, i.e. albums with the same name as the artist. Write a query that will highlight all eponymous albums as well as non-eponymous albums.

```python
%%sql

SELECT
    ar.Name,
    a.Title
FROM
    artists ar
LEFT JOIN
    albums AS a
    ON ar.Name = a.Title

UNION ALL

SELECT
    ar.Name,
    a.Title
FROM
    albums AS a
LEFT JOIN
    artists AS ar
    ON ar.Name = a.Title
```

## JOINS & SET OPERATIONS

####
Using the Northwind database, as the new sales analyst, you've been asked to prepare a sales report for the last quarter. Your report should include:

1. Customer names and their countries.

2. The total amount spent by each customer.

3. The list of products each customer purchased.

4. The total quantity of each product purchased by each customer.

Only include customers who have spent more than 5000 in the last quarter.

```python
%sql sqlite:///northwind.db
```

```python
%%sql


SELECT
    customer.ContactName,
    customer.Country,
    product.ProductName,
    SUM(orderDetails.Quantity) AS TotalUnitsSold,
    SUM(orderDetails.Quantity * orderDetails.UnitPrice) AS TotalSalesAmount
FROM
    Customers AS customer
JOIN
    Orders AS orders
    ON customer.CustomerID = orders.CustomerID
JOIN
    OrderDetails AS orderDetails
    ON orders.OrderID = orderDetails.OrderID
JOIN
    Products AS product
    ON orderDetails.ProductID = product.ProductID
WHERE
    orders.OrderDate BETWEEN '1996-08-01' AND '1998-01-30'
GROUP BY
    customer.ContactName,
    customer.Country,
    product.ProductName
HAVING
    SUM(orderDetails.Quantity * orderDetails.UnitPrice) > 5000;
```
